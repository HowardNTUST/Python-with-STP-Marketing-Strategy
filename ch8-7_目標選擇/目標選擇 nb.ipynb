```python
# 載入stopwords
with open('stopwords.txt', encoding = 'utf8') as f:
    stopwords = f.readlines()
stopwords= [w.replace('\n', '') for w in stopwords]
stopwords= [w.replace(' ', '') for w in stopwords]
stopwords.append('\n')
stopwords.append('\n   \n')
stopwords.append('\x0b')
```

```python
# 將stopwords（list）加入removeword
# hint extend
stopwords.extend(removeword)
stopwords
```

```python
#--- 移除無意義字元列
for word in stopwords:
    underwearData['內文'] = underwearData['內文'].str.replace(word,'')

#--- NA取代為空值
underwearData['內文'] = underwearData['內文'].fillna('')

# 將內文融合，變成一組大字串
ptt_string = underwearData['內文'].sum() 


#--- 切詞
list(jieba.cut('我非常喜歡Howard'))
words = list(jieba.cut(ptt_string))


#------ 計算關鍵字數量-------

# 先將words（data type爲list），轉換成data frame
tag = pd.DataFrame({'tag':words})
tag['tag'].value_counts()
```
